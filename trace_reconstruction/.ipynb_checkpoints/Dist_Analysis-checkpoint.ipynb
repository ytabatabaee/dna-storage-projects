```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_distance_histogram(dtype, hist, params):
    plt.xlabel(dtype + ' distance')
    plt.ylabel('test count')
    plt.plot(range(len(hist)), hist, label=params)
    plt.title('distribution of ' + dtype + ' distances')
    plt.legend()
    # plt.show()
```

```python
plot_distance_histogram('edit', edit_dist, 'L = 100, p = 0.15, N = 10, #T = 10000')
plt.show()
```

```python
plot_distance_histogram('hamming', hamming_dist, 'L = 100, p = 0.15, N = 10, #T = 10000')
plt.show()
```

```python
def plot_distance_histogram(ed, hd, params, log=''):
    plt.xlabel('distance')
    plt.ylabel('test count ' + log)
    plt.plot(20'''range(len(ed))''', (ed), label='edit, avg=0.027')
    plt.plot(20'''range(len(hd))''', (hd), label='hamming, avg=0.175')
    plt.title(params)
    plt.legend()
```

```python
plot_distance_histogram(edit_dist, hamming_dist, 'Baseline, L = 100, p = 0.15, N = 10, #T = 10000')#, '(log10 scale)')
plt.savefig('params.pdf')
```
